```python
from sklearn.datasets import load_breast_cancer #loaded dataset
breast = load_breast_cancer()
breast_data = breast.data
print(breast)
print()
print(breast_data) #dataset print
print()
breast_data.shape #data shape
print()
breast_labels = breast.target
print(breast_labels)  #labels
breast_labels.shape  #labels shape


```

```python
import numpy as np
labels = np.reshape(breast_labels,(569,1)) #reshape
final_breast_data = np.concatenate([breast_data,labels],axis=1)  #concatenate axis = 1
final_breast_data.shape #shape 
print(labels) 
print()
print(final_breast_data) 

```

```python
import pandas as pd
breast_dataset = pd.DataFrame(final_breast_data) #data frame
print(breast_dataset)
print()
features = breast.feature_names
print(features)
```

```python
features_labels = np.append(features,'label')
breast_dataset.columns = features_labels
print(features_labels)
print()
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
breast_dataset.tail()

```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
print(feat_cols)
```

```python
normalised_breast = pd.DataFrame(x,columns=feat_cols)
print(normalised_breast)
```

```python
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
print(pca_breast)
print()

print(principalComponents_breast)
print()

print(principal_breast_Df)
print()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
