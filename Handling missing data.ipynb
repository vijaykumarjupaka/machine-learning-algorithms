```python
# importing pandas module
import pandas as pd
   
# making data frame from csv file
data = pd.read_csv("weatherAUS.csv")
   
# making new data frame with dropped NA values
new_data = data.dropna(axis = 0, how ='any')
   
new_data
```

```python
print("Old data frame length:", len(data))
print("New data frame length:", len(new_data))
print("Number of rows with at least 1 NA value: ", (len(data)-len(new_data)))
```

```python
#Impute / Replace Missing Values with Mean
data.fillna(data.mean())
```

```python
#Impute / Replace Missing Values with Median
data.fillna(data.median())
```

```python
#Impute / Replace Missing Values with Mode
data['Evaporation'] = data['Evaporation'].fillna(data['Evaporation'].mode()[0])
data['Evaporation'] 
```
